```python
import numpy as np
import random
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import LabelEncoder, StandardScaler

def preprocess_features(data):
   

    feature_maps = {
        'label_encoders': {},
        'scaler': StandardScaler()
    }

    processed_data = data.copy()

    # Handle categorical features
    for column in processed_data.columns:
        if processed_data[column].dtype == 'object' or pd.api.types.is_categorical_dtype(processed_data[column]):
            le = LabelEncoder()
            processed_data[column] = le.fit_transform(processed_data[column])
            feature_maps['label_encoders'][column] = le

    # Scale all features
    processed_data = feature_maps['scaler'].fit_transform(processed_data)

    return processed_data, feature_maps

def canopy_feature_selection(data, T1, T2):
   
    features = np.array(range(data.shape[1]))  # Feature indices
    selected_features = []

    while len(features) > 0:
        # Pick a random feature as the canopy center
        center_idx = random.choice(features)
        selected_features.append(center_idx)

        # Compute distances between features
        distances = euclidean_distances(data.T[features], data.T[[center_idx]]).flatten()

        # Remove features within T2 (redundant ones)
        features = features[distances > T2]

    return selected_features

# Example Usage
if __name__ == "__main__":
    # Load your dataset
    df = pd.read_csv("dataset_phishing.csv")  # Your dataset with categorical features

    # Separate target variable if exists
    if 'status' in df.columns:
        X = df.drop(columns=['status'])
        y = df['status']
    else:
        X = df

    # Preprocess the features
    processed_data, feature_maps = preprocess_features(X)

    # Set thresholds (T1 > T2)
    T1, T2 = 0.8, 0.5  # Adjust based on feature correlations

    # Select important features
    selected_indices = canopy_feature_selection(processed_data, T1, T2)

    # Get selected feature names
    selected_columns = X.columns[selected_indices].tolist()

    # Create a new dataframe with selected features
    df_selected = X[selected_columns]

    print("Selected Features:", selected_columns)
    print("New Data Shape:", df_selected.shape)

    # Print feature types for verification
    print("\nFeature Types:")
    for col in selected_columns:
        print(f"{col}: {X[col].dtype}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 1. Define your evaluate_model function
def evaluate_model(model_name, y_test, y_pred):
    # Modify pos_label if your positive class is something else
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label="phishing")
    recall = recall_score(y_test, y_pred, pos_label="phishing")
    f1 = f1_score(y_test, y_pred, pos_label="phishing")

    print(f"\nMetrics for {model_name}:")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")

    
    return {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    }
model_metrics = {}
```

```python
# Evaluate Logistic Regression
y_pred_lr = model_results["Logistic Regression"]["Model"].predict(X_test)
model_metrics["Logistic Regression"] = evaluate_model("Logistic Regression", y_test, y_pred_lr)

# Evaluate Random Forest
y_pred_rf = model_results["Random Forest"]["Model"].predict(X_test)
model_metrics["Random Forest"] = evaluate_model("Random Forest", y_test, y_pred_rf)

# Evaluate K-Nearest Neighbors
y_pred_knn = model_results["K-Nearest Neighbors"]["Model"].predict(X_test)
model_metrics["K-Nearest Neighbors"] = evaluate_model("K-Nearest Neighbors", y_test, y_pred_knn)

# Evaluate Gradient Boosting
y_pred_gbm = model_results["Gradient Boosting"]["Model"].predict(X_test)
model_metrics["Gradient Boosting"] = evaluate_model("Gradient Boosting", y_test, y_pred_gbm)

# Evaluate Decision Tree
y_pred_dt = model_results["Decision Tree"]["Model"].predict(X_test)
model_metrics["Decision Tree"] = evaluate_model("Decision Tree", y_test, y_pred_dt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    confusion_matrix, roc_curve, auc, classification_report, precision_recall_curve
)

#  Split Data (Assuming X_pca is your PCA-transformed data)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)

# Define Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}

# Dictionary to Store Results
model_results = {}

# Train & Evaluate Each Model Separately
for model_name, model in models.items():
    print(f"\nTraining {model_name}...\n")
    
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # Probability for positive class

    # Compute Metrics
    cm = confusion_matrix(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_proba, pos_label="phishing")
    auc_score = auc(fpr, tpr)
    precision, recall, _ = precision_recall_curve(y_test, y_proba, pos_label="phishing")

    # Store results
    model_results[model_name] = {
        "Model": model,
        "Confusion Matrix": cm,
        "ROC Curve": (fpr, tpr),
        "AUC": auc_score,
        "Precision-Recall Curve": (precision, recall),
        "Classification Report": classification_report(y_test, y_pred, output_dict=True)
    }

    #  PLOT CONFUSION MATRIX
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Legitimate", "Phishing"],
                yticklabels=["Legitimate", "Phishing"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    # PLOT ROC CURVE
    plt.figure(figsize=(5, 4))
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {auc_score:.2f})")
    plt.plot([0, 1], [0, 1], color="gray", linestyle="--")  # Diagonal reference line
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name}")
    plt.legend()
    plt.show()

    #  PLOT PRECISION-RECALL CURVE
    plt.figure(figsize=(5, 4))
    plt.plot(recall, precision, label=f"{model_name}")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Precision-Recall Curve - {model_name}")
    plt.legend()
    plt.show()

# Print Summary of AUC Scores
print("\nAUC Scores for Each Model:")
for model_name, results in model_results.items():
    print(f"{model_name}: {results['AUC']:.2f}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, BatchNormalization, Flatten, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("dataset_phishing.csv")  # Replace with your dataset

# Separate features and target variable
if 'status' in df.columns:
    X = df.drop(columns=['status'])
    y = df['status']
else:
    X = df

# Encode categorical columns
label_encoders = {}
for column in X.columns:
    if X[column].dtype == 'object':  
        le = LabelEncoder()
        X[column] = le.fit_transform(X[column])
        label_encoders[column] = le  

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA, keeping 95% variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Reshape data for CNN (Adding extra dimension for Conv1D)
X_train_cnn = np.array(X_train, dtype=np.float32)[..., None]  
X_test_cnn = np.array(X_test, dtype=np.float32)[..., None]

# Convert y_train and y_test to float32
y_train = np.array(y_train, dtype=np.float32)
y_test = np.array(y_test, dtype=np.float32)

# Define Improved CNN Model
cnn_model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation="relu", kernel_regularizer=l2(0.001), input_shape=(X_train_cnn.shape[1], 1)),
    BatchNormalization(),  # Normalize feature maps
    Dropout(0.5),  # Dropout to reduce overfitting

    Conv1D(filters=32, kernel_size=3, activation="relu", kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.5),

    Flatten(),
    Dense(128, activation="relu", kernel_regularizer=l2(0.001)),
    Dropout(0.5),
    Dense(1, activation="sigmoid")  # Binary classification
])

# Compile model
cnn_model.compile(optimizer=Adam(learning_rate=0.0005), loss="binary_crossentropy", metrics=["accuracy"])

# Train model with validation set
history = cnn_model.fit(X_train_cnn, y_train, validation_data=(X_test_cnn, y_test), epochs=50, batch_size=32)

# Evaluate model
test_loss, test_accuracy = cnn_model.evaluate(X_test_cnn, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

def evaluate_and_compare_models(X_test, y_test, models):
    """
    Evaluate and compare multiple trained models including LightGBM
    """
    # Ensure y_test is flattened
    y_test = y_test.ravel()
    
    # Prepare test data for different architectures
    X_test_cnn = X_test.reshape(-1, X_test.shape[1], 1)
    X_test_rnn = X_test.reshape(-1, X_test.shape[1], 1)
    
    # Dictionary to store all results
    results = {}
    
    # Function to process predictions consistently
    def process_predictions(y_pred):
        if isinstance(y_pred, list):
            y_pred = np.array(y_pred)
        if y_pred.ndim > 1:
            if y_pred.shape[1] > 1:  # Multi-class
                return np.argmax(y_pred, axis=1)
            else:  # Binary
                return (y_pred.ravel() > 0.5).astype(int)
        return y_pred.ravel()
    
    # Evaluate each model
    model_configs = {
        'CNN': (models['CNN'], X_test_cnn, 'nn'),
        'RNN': (models['RNN'], X_test_rnn, 'nn'),
        'MLP': (models['MLP'], X_test, 'nn'),
        'LightGBM': (models['LightGBM'], X_test, 'gbm')
    }
    
    for model_name, (model, X, model_type) in model_configs.items():
        try:
            # Get predictions based on model type
            if model_type == 'nn':
                y_pred = model.predict(X, verbose=0)
            else:  # LightGBM
                y_pred = model.predict(X)
                
            # Process predictions to ensure consistent format
            y_pred_processed = process_predictions(y_pred)
            
            # Calculate metrics
            results[model_name] = {
                'Accuracy': accuracy_score(y_test, y_pred_processed),
                'Precision': precision_score(y_test, y_pred_processed, average='binary', zero_division=1),
                'Recall': recall_score(y_test, y_pred_processed, average='binary', zero_division=1),
                'F1 Score': f1_score(y_test, y_pred_processed, average='binary', zero_division=1)
            }
            print(f"\n{model_name} evaluation completed successfully")
            
        except Exception as e:
            print(f"\nError evaluating {model_name}: {str(e)}")
            results[model_name] = {
                'Accuracy': None,
                'Precision': None,
                'Recall': None,
                'F1 Score': None
            }
    
    # Convert results to DataFrame
    df_results = pd.DataFrame(results).T
    
    # Print results
    print("\nModel Performance Metrics:")
    print(df_results.round(4))
    
    # Remove any None values for plotting
    df_plot = df_results.dropna()
    
    if not df_plot.empty:
        # Bar plot
        plt.figure(figsize=(12, 6))
        df_plot.plot(kind='bar', width=0.8)
        plt.title('Model Performance Comparison')
        plt.xlabel('Models')
        plt.ylabel('Score')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
        
        # Heatmap
        plt.figure(figsize=(10, 6))
        sns.heatmap(df_plot, annot=True, cmap='YlOrRd', fmt='.4f')
        plt.title('Model Performance Heatmap')
        plt.tight_layout()
        plt.show()
        
        # Find best model for each metric
        best_models = df_plot.idxmax()
        print("\nBest performing models:")
        for metric, model in best_models.items():
            print(f"{metric}: {model} ({df_plot.loc[model, metric]:.4f})")
    
    return df_results

# Create dictionary of your trained models
models = {
    'CNN': cnn_model,
    'RNN': rnn_model,
    'MLP': mlp_model,
    'LightGBM': lgb_model  # Your trained LightGBM model
}

# Run evaluation and comparison
results_df = evaluate_and_compare_models(X_test, y_test, models)
```
